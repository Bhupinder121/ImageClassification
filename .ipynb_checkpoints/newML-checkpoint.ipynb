```python
import tensorflow as tf
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
standard_dir = "/content/drive/MyDrive/ImageClassification/"
```

```python
train = "archive/train"
print(os.listdir(train))
```

```python
data = tf.keras.utils.image_dataset_from_directory(train)
data
data_iterator = data.as_numpy_iterator()
```

```python
batch = data.as_numpy_iterator().next()
```

```python
batch[1]
```

```python
max = len(batch[1])
max = 5
fig, ax = plt.subplots(ncols=max, figsize=(20,20))
for idx, img in enumerate(batch[0][:max]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text([batch[1][idx]])
```

```python
data = data.map(lambda x,y: (x/255, y))
```

```python
len(data)
```

```python
train_size = int(len(data)*0.8)
val_size = int(len(data)*0.2)+1


train_size + val_size
```

```python
batch[0].max()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile("adam", loss=tf.losses.CategoricalCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir=standard_dir+'logs'
logdir
```

```python
callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
training_data = data.take(train_size)
val_data = data.skip(train_size).take(val_size)
```

```python
len(val_data)+len(training_data)
```

```python
hist = model.fit(training_data, epochs=20, validation_data=val_data, callbacks=[callback])
```
